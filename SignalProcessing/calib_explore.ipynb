```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
```

```python
center = (0, 0)  # Center coordinates
range_value = 103  # Range in meters
angle_range = 75  # Degrees on either side of the center (total 150 degrees)

# Convert angle to radians
angle_range_rad = angle_range * np.pi / 180

# Starting and ending angles for the arc (considering both sides)
start_angle = -angle_range_rad
end_angle = angle_range_rad

# Width (twice the angle range) and height (range value)
width = 2 * angle_range_rad
height = range_value

# Create an arc patch with specified center, width, height, angles, and color
arc = Arc(xy=center, 
          width=width,
          height=height, 
          angle=0, 
          theta1=start_angle, 
          theta2=end_angle, 
          color='lightblue',
          linewidth=2)
```

```python
fig, ax = plt.subplots()
ax.add_patch(arc)
ax.set_aspect('equal')  # Set equal aspect ratio for circular plot

# Optional: Plot a circle at the origin for reference
theta = np.linspace(0, 2*np.pi, 200)
radius = range_value
a = radius * np.cos( theta )
b = radius * np.sin( theta )
plt.plot(a, b, color='gray', linestyle='--', label='Maximum Range')

# Labels and title (optional)
plt.xlabel('X-Coordinate (Meters)')
plt.ylabel('Y-Coordinate (Meters)')
plt.title('RADAR Cone (75 Degrees Each Side)')

# Legend (optional)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

# Define parameters with increased scale factor
scale_factor = 1  # Increase scale by a factor of 2
center = (0, 0)
range_value = 103 * scale_factor
angle_range = 75  # Degrees on either side (total 150 degrees)

# Convert angle to radians
angle_range_rad = angle_range * np.pi / 180

# Starting and ending angles for the arc (considering both sides)
start_angle = -angle_range_rad
end_angle = angle_range_rad

# Width (twice the angle range) and height (range value)
width = 2 * angle_range_rad
height = range_value

# Create arc patch
arc = Arc(xy=center,
          width=width,
          height=height,
          angle=0,
          theta1=start_angle,
          theta2=end_angle,
          color='lightblue',
          linewidth=2)

# Create plot with swapped axes
fig, ax = plt.subplots(figsize=(6, 6))
ax.add_patch(arc)

# Set aspect ratio to 'equal' and adjust limits for swapped axes
ax.set_aspect('equal')
ax.set_xlim(-range_value * scale_factor, range_value * scale_factor)
ax.set_ylim(-angle_range_rad, angle_range_rad)  # Swap y-axis limits for correct scaling

# Optional circle for reference (scaled)
theta = np.linspace(-angle_range_rad, angle_range_rad, 200)
radius = range_value
a = radius * np.cos(theta)
b = radius * np.sin(theta)
plt.plot(a, b, color='gray', linestyle='--', label='Maximum Range')

# Connect lines to +/- 75 degrees (scaled)
x_line_1 = np.linspace(-range_value * scale_factor, range_value * scale_factor, 2)
y_line_1 = np.full_like(x_line_1, angle_range_rad * scale_factor)
x_line_2 = np.linspace(-range_value * scale_factor, range_value * scale_factor, 2)
y_line_2 = np.full_like(x_line_1, -angle_range_rad * scale_factor)
plt.plot(x_line_1, y_line_1, color='red', linewidth=1, linestyle='-')
plt.plot(x_line_2, y_line_2, color='red', linewidth=1, linestyle='-')

# Labels and title (optional)
plt.xlabel('Y-Coordinate (Scaled)')  # Swapped label
plt.ylabel('X-Coordinate (Degrees)')  # Swapped label
plt.title('RADAR Cone (75 Degrees Each Side)')

# Legend (optional)
plt.legend()
plt.grid(True)
plt.show()
```

```python
image = np.random.rand(512, 751)
```

```python
import numpy as np

# Assuming your image is a NumPy array named 'image'
shape = image.shape  # Get the image shape
center = (0, 376)  # Center coordinates

# Define threshold angles in radians
lower_angle = -10 * np.pi / 36
upper_angle = 10 * np.pi / 36

# Calculate maximum distance from center within the valid angle range
max_dist = 512  # Consider only half the width for positive angles

# Create a mask to select valid pixels
mask = np.zeros(shape, dtype=bool)
y, x = np.ogrid[:shape[0], :shape[1]]
distance_from_center = np.sqrt((y - center[0])**2 + (x - center[1])**2)
mask = np.logical_and(distance_from_center <= max_dist, np.arctan2(y - center[0], x - center[1]) >= lower_angle)

# Apply the mask to keep only desired values
filtered_image = np.copy(image)  # Copy the original image
filtered_image[~mask] = 0  
# Now 'filtered_image' contains only the values within the specified angle range.
```
